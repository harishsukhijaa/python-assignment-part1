```python
# Name: Harish Sukhija
# Student ID: bitsom_ba_2511153
# Assignment: Python Basics & Control Flow

print("Task 1 - Data Parsing & Profile Cleaning\n")

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]


name_for_roll_103 = ""


for student in raw_students:


    clean_name = student["name"].strip().title()


    clean_roll = int(student["roll"])


    marks_list_strings = student["marks_str"].split(", ")
    clean_marks = []
    for m in marks_list_strings:
        clean_marks.append(int(m))


    is_valid = True
    for word in clean_name.split():
        if not word.isalpha():
            is_valid = False
            break

    if is_valid:
        valid_status = "✓ Valid name"
    else:
        valid_status = "✗ Invalid name"

    if clean_roll == 103:
        name_for_roll_103 = clean_name


    print("================================")
    print(f"Student : {clean_name} {valid_status}")
    print(f"Roll No : {clean_roll}")
    print(f"Marks   : {clean_marks}")
    print("================================")
    print()


print(f"Roll 103 Name in ALL CAPS: {name_for_roll_103.upper()}")
print(f"Roll 103 Name in lowercase: {name_for_roll_103.lower()}")
print("\n" + "="*40 + "\n")
```

```python
# ==========================================
# Task 2: Marks Analysis Using Loops
# ==========================================
print("Task 2 - Marks Analysis Using Loops & Conditionals\n")

student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

total_marks = 0
highest_score = -1
highest_subject = ""
lowest_score = 101
lowest_subject = ""

print(f"Student: {student_name}")
print("-" * 30)


for i in range(len(subjects)):
    sub = subjects[i]
    m = marks[i]
    total_marks += m

    # Grade logic (If-Else)
    if m >= 90 and m <= 100:
        grade = "A+"
    elif m >= 80 and m <= 89:
        grade = "A"
    elif m >= 70 and m <= 79:
        grade = "B"
    elif m >= 60 and m <= 69:
        grade = "C"
    else:
        grade = "F"

    # Highest score check
    if m > highest_score:
        highest_score = m
        highest_subject = sub

    # Lowest score check
    if m < lowest_score:
        lowest_score = m
        lowest_subject = sub

    print(f"{sub}: {m} (Grade: {grade})")

average_marks = round(total_marks / len(subjects), 2)

print("-" * 30)
print(f"Total marks: {total_marks}")
print(f"Average marks: {average_marks}")
print(f"Highest scoring subject: {highest_subject} ({highest_score})")
print(f"Lowest scoring subject: {lowest_subject} ({lowest_score})\n")


# 2. While loop: Interactive marks entry system
print("--- Add New Subjects ---")
new_subjects_added = 0

while True:
    new_sub = input("Enter subject name (or type 'done' to stop): ")


    if new_sub.lower() == 'done':
        break

    new_marks_str = input(f"Enter marks for {new_sub} (0-100): ")


    if not new_marks_str.isdigit():
        print("Warning: Please enter valid numeric marks!")
        continue

    new_m = int(new_marks_str)


    if new_m < 0 or new_m > 100:
        print("Warning: Marks must be between 0 and 100!")
        continue


    subjects.append(new_sub)
    marks.append(new_m)
    new_subjects_added += 1
    print(f"Successfully added {new_sub} with {new_m} marks.\n")


final_total = sum(marks)
final_average = round(final_total / len(marks), 2)

print("\n--- Final Update ---")
print(f"Total new subjects added: {new_subjects_added}")
print(f"Updated average marks: {final_average}")
```

```python
# ==========================================
# Task 3: Class Performance Summary
# ==========================================
print("Task 3 - Class Performance Summary")
print("-" * 45)

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print(f"{'Name':<18} | {'Average':<7} | {'Status'}")
print("-" * 40)

passed_count = 0
failed_count = 0
top_student_name = ""
highest_avg = -1
sum_of_averages = 0


for student in class_data:
    name = student[0]
    marks_list = student[1]


    avg = round(sum(marks_list) / len(marks_list), 2)
    sum_of_averages += avg


    if avg >= 60:
        status = "Pass"
        passed_count += 1
    else:
        status = "Fail"
        failed_count += 1

    if avg > highest_avg:
        highest_avg = avg
        top_student_name = name

    print(f"{name:<18} |  {avg:<5.2f}  | {status}")

class_total_avg = round(sum_of_averages / len(class_data), 2)

print("\n--- Final Summary ---")
print(f"Number of students who passed: {passed_count}")
print(f"Number of students who failed: {failed_count}")
print(f"Class topper: {top_student_name} ({highest_avg})")
print(f"Class average: {class_total_avg}\n\n")


# ==========================================
# Task 4: String Manipulation Utility
# ==========================================
print("Task 4 - String Manipulation Utility")
print("-" * 45)

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "


clean_essay = essay.strip()
print("1. Stripped Essay:")
print(clean_essay)


title_case_essay = clean_essay.title()
print("\n2. Title Case Essay:")
print(title_case_essay)


python_count = clean_essay.count("python")
print(f"\n3. The word 'python' appears {python_count} times.")


replaced_essay = clean_essay.replace("python", "Python 🐍")
print("\n4. Replaced Essay:")
print(replaced_essay)

sentences_list = clean_essay.split(". ")
print("\n5. List of Sentences:")
print(sentences_list)

print("\n6. Numbered Sentences:")

for count, sentence in enumerate(sentences_list, 1):

    if not sentence.endswith("."):
        sentence = sentence + "."
    print(f"{count}. {sentence}")
```
